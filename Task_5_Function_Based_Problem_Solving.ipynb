```python
"""Problem 1: Social Media – Post Engagement Analyzer
Problem Statement
Create a function to analyze engagement on a social media post.
Rules
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement
"""
#function to analyze engagement on a social media post
def engagement_analyze(likes_list):
    #calculate total likes
    total_likes = 0
    for like in likes_list:
        total_likes += like
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
#output
    print("Total Likes:", total_likes)
    print("Post Status:", status)
#input
engagement_analyze([200, 300, 250, 300])
```

```python
"""Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement
Create a function that analyzes seasonal rainfall data.
Rules
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall
"""
#function that analyzes seasonal rainfall data
def rainfall_check(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain
#to find avg
    average = int(total / len(rainfall_list))
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
#output
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)
#input
rainfall_check([70, 75, 68, 75], 70)
```

```python
"""Problem 4: Social Media – Duplicate Account Detection
Problem Statement
Create a function to detect duplicate usernames in a system.
Rules
If duplicates exist → Flag accounts
Else → Accounts are unique
"""
#function to detect duplicate usernames in a system
def duplicate_account(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
#input
duplicate_account(["user1", "user2", "user3", "user1"])
```

```python
"""Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement
Create a function that filters crop prices above a premium threshold.
Rules
Price > ₹2000 → Premium Crop
Else → Regular Crop
"""
#function that filters crop prices above a premium threshold
def premium_crop(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
#output
    print("Premium Crops:", premium_crops)
#input
premium_crop([1500, 2500, 3200, 1800])
```

```python
"""Problem 7: System Monitoring – Application Health Checker
Problem Statement
Create a function to evaluate application health based on error count.
Rules
Errors = 0 → Healthy
Errors ≤ 5 → Minor Issues
Errors > 5 → Critical Issues
"""
#function to evaluate application health based on error count
def application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
#output
    print("Error Count:", errors)
    print("System Status:", status)
#input
application_health(7)
```

```python
"""Problem 8: Banking – Daily Transaction Limit Checker
Problem Statement
Create a function to validate daily transaction amount.
Rules
Daily limit = ₹50,000
Amount within limit → Approved
Amount exceeds limit → Rejected
"""
#function to validate daily transaction amount
def transaction_limit_check(amount):
    daily_limit = 50000
#logic
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
#output
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
#input
transaction_limit_check(60000)
```

```python
"""Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
Rules
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible
"""
#function to calculate attendance percentage and check exam eligibility
def attendance_check(attendance_list):
    present_count = 0
#logic
    for day in attendance_list:
        if day == "P":
            present_count += 1
    percentage = (present_count / len(attendance_list)) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
#output
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)
#input
attendance_check(["P", "P", "A", "P", "P"])
```

```python
"""Problem 10: Smart Electricity Bill Analyzer
Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
Billing Rules
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit
Usage Classification
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage
"""
#function that calculates the electricity bill and classifies usage level based on units consumed
def electricity_bill(units):
    bill = 0
#logic
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
#output
    print("Total Bill:", bill)
    print("Usage Level:", usage)
# input
electricity_bill(250)
```
