```python
"""Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
"""
#class to control a smart light
class Light:
  #to initialise name of the light and status of the light
  def __init__(self, nameoflight):
    self.nameoflight = nameoflight
    self.lightstatus = "OFF"
  #method for light on
  def lighton(self):
    self.lightstatus = "ON"
  #method for light off
  def lightoff(self):
    self.lightstatus = "OFF"
  #method to display light status
  def displaystatus(self):
    print(f"{self.nameoflight} is {self.lightstatus}")
  #input
light1 = Light("Bedroom Light")
light1.lighton()
light1.displaystatus()
```

```python
"""Problem 2: Employee ID Card System
 Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.
"""
#class that stores employee name, ID, and department and displays ID card details
class employee:
  #initialise emplyee name, id, department
  def __init__(self, emp_name, emp_id, emp_dept):
    self.emp_name = emp_name
    self.emp_id = emp_id
    self.emp_dept = emp_dept
  #method to display id card details
  def display_id(self):
    print("Employee ID Card")
    print(f"Employee Name: {self.emp_name}")
    print(f"Employee ID: {self.emp_id}")
    print(f"Department: {self.emp_dept}")
  #input
emp1 = employee("Rahul", "EMP101", "IT")
emp1.display_id()

```

```python
"""Problem 3: Mobile Contact Record
Problem Statement
Create a class to store contact name and phone number and display contact information.
"""
#class to store contact name and phone number and display contact information
class contact:
  #to initialise name and number
  def __init__(self, name, phonenumber):
    self.name = name
    self.phonenumber = phonenumber
    #method to display contact info
  def display_contact_info(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Phone: {self.phonenumber}")
#input
contact1 = contact("Anita", "9876543210")
contact1.display_contact_info()
```

```python
"""Problem 4: Product Price Tag Generator
Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
"""
#class that stores product name and price
class product:
  #initialising product name and price
    def __init__(self,prd_name, prd_price):
        self.prd_name = prd_name
        self.prd_price = prd_price
    # to print a formatted price tag
    def print_pricetag(self):
        print(f"Product: {self.prd_name}")
        print(f"Price: ₹{self.prd_price}")
#input
product1 = product("Headphones", "2499")
product1.print_pricetag()
```

```python
"""Problem 5: Movie Rating Display System
Problem Statement
Create a class that stores movie name and rating and displays movie details.
Real-Time Use
"""
#class that stores movie name and rating
class movie:
  #initialise movie name and ratings
    def __init__(self,movie_name, movie_rating):
      self.movie_name = movie_name
      self.movie_rating = movie_rating
    #to display movie details
    def display_movie_details(self):
      print(f"Movie: {self.movie_name}")
      print(f"Rating: {self.movie_rating}")
#input
movie1 = movie("Inception","4.8 / 5")
movie1.display_movie_details()
```

```python
"""Problem 6: Delivery Address Manager
Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.
"""
#class that stores customer name and delivery address
class delivery:
  #initialise customer name and delivery address
  def __init__(self, customer_name, delivery_address):
    self.customer_name = customer_name
    self.delivery_address = delivery_address
  #to print delivery details
  def print_delivery_info(self):
    print("Delivery Details")
    print(f"Customer: {self.customer_name}")
    print(f"Address: {self.delivery_address}")
#input
delivery1 = delivery("Suman","Hyderabad")
delivery1.print_delivery_info()
```
