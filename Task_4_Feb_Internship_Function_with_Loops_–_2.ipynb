```python
"""Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid
"""
# Function creation
def recharge_valid(recharge_amt, plan):
    # Recharge amount must be ≥ ₹50
    if recharge_amt < 50:
        return "Recharge unsuccessful! Enter valid amount"
    # Valid plans: 199, 299, 399, 599
    if plan not in [199, 299, 399, 599]:
        return "Recharge unsuccessful! Enter valid plan"
    # Recharge successful
    return "Recharge Successful"
# While loop for retry
while True:
    recharge_amt = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan (199/299/399/599): "))
    result = recharge_valid(recharge_amt, plan)
    print(result)
    if result == "Recharge Successful":
        break
    else:
        print("Please try again.\n")
```

```python
"""Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise"""
# Function to check stock levels
def check_stock(products):
  for item, quantity in products.items():
    if quantity < 15:
      print(f"Reorder alert: {item} stock is {quantity}")
    else:
      print(f"{item} stock is OK")
# Dictionary of products and stock quantities
products = {
  "Pens": 12, "Erasers": 34, "Pencils": 54, "Sharpeners": 10 }
# Function call
check_stock(products)
```

```python
""" Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average ≥ 50 else Fail"""
# Function to process student marks
def process_students_marks(marks):
  total = 0
# Loop to calculate total
  for mark in marks:
    total += mark
  avg = total / len(marks)
  if avg >= 50:
   return f"Average = {avg:.2f} → PASS"
  else:
   return f"Average = {avg:.2f} → FAIL"
# Taking input from user
marks = []
n = int(input("Enter number of subjects: "))
for i in range(n):
    mark = int(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)
# Function call
result = process_students_marks(marks)
print(result)
```

```python
"""Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50
₹12 per km
Peak hour → 25% extra
Allow retry using while loop
"""
# Function to calculate fare
def cab_fare_calculator(distance, peak):
    base_fare = 50
    rate_per_km = 12
    peak_hour_rate = 0.25
    total_fare = base_fare + (distance * rate_per_km)
    # Peak hour → 25% extra
    if peak.lower() == "yes":
        total_fare = total_fare + (peak_hour_rate * total_fare)
    return total_fare
# Retry system using while loop
while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))
        peak = input("Is it peak hour? (yes/no): ")
        fare = cab_fare_calculator(distance, peak)
        print(f"Total Cab Fare = ₹{fare:.2f}")
        break   # Exit after successful calculation
    except ValueError:
        print("Invalid input. Please enter correct values.\n")
```

```python
"""Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if ≥ 75% else Not Eligible
"""
# Function to check attendance eligibility
def attendance_elig(attendance):
  count_of_present = 0
   # Loop to count present days
  for day in attendance:
    if day.upper() == 'P':
      count_of_present += 1
  present_percent = (count_of_present / len(attendance)) * 100
  if present_percent >= 75:
        return f"Attendance = {present_percent:.2f}% → Eligible"
  else:
        return f"Attendance = {present_percent:.2f}% → Not Eligible"
  # Taking input from user
attendance = []
n = int(input("Enter total working days: "))
for x in range(n):
    day = input(f"Enter attendance for day {x+1} (P/A): ")
    attendance.append(day)
# Function call
result = attendance_elig(attendance)
print(result)



```

```python
"""Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)
"""
# Function to check password strength
def password_strength_check(password):
  has_digit = False
  has_special = False
  special_chars = "@#$"
  #Minimum length
  if len(password) < 8:
    return "Weak (Minimum 8 characters required)"
     # Loop over string
  for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
  if has_digit and has_special:
        return "Strong Password"
  else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"


# Taking input
password = input("Enter your password: ")
result = password_strength_check(password)
print(result)
```
