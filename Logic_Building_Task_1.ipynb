```python
"""1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"
Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"
"""
#given data
username = "admin"
password = "1234"

#input from user
userName = input("enter your username: ")
passWord = input("enter the correct password: ")

#logic
if userName == username and passWord == password:
  print("Login Successful")

else:
  print("Invalid Credentials")

```

```python
"""Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly
"""

#given data
marks = [45, 78, 90, 33, 60]

#number of passed and failed students initialised
passed_students = 0
failed_students = 0

#logic
for i in marks:
  if i>=50:
    passed_students+=1
  else:
    failed_students+=1

#final result
print("total students passed are " ,passed_students)
print("total students failed are " ,failed_students)
```

```python
"""3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list
"""
#given
names = [" Alice ", "bob", " CHARLIE "]

#to store names after cleaning
clean_names = []

#removing space
for i in range(len(names)):
  names[i] = names[i].strip()

#converting each to lower case
  names[i] = names[i].lower()

#to print cleaned list
print("The list after cleaning: " ,names)
```
