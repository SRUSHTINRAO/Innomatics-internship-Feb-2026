```python
"""Problem Statement 1:Smart Parking Lot Management System
Design a function to manage a smart parking lot.
The system should:
Accept vehicle entry and exit logs
Calculate total parked vehicles
Identify peak parking usage
Alert if parking exceeds capacity
"""
def smart_parking(capacity, exit_logs):
    parked_vehicles = 0
    peak_usage = 0
#function to manage a smart parking lot
    for x in exit_logs:
        x = x.upper()   # convert input to uppercase
        if x == "IN":
            parked_vehicles += 1
        elif x == "OUT":
            if parked_vehicles > 0:
                parked_vehicles -= 1
        if parked_vehicles > peak_usage:
            peak_usage = parked_vehicles
    print("Currently Parked Vehicles:", parked_vehicles)
#to check capacity
    if parked_vehicles > capacity:
        print("Parking exceeds capacity")
    else:
        print("Parking Status: Available")
# Input
capacity = int(input("Enter the capacity of the parking lot: "))
exit_logs = input("Enter logs (IN/OUT): ").split()
smart_parking(capacity, exit_logs)
```

```python
""" Problem Statement 2:Online Food Delivery Time Estimator
Create a function that estimates delivery time based on:
Distance (km)
Weather condition
Traffic level
Apply delays dynamically and display final ETA."""
#function that estimates delivery time based on
def delivery_time(distance, weather, traffic):
  # Base delivery time
  ETA = distance * 5
  # Traffic delay
  if traffic == "high":
    ETA += 20
  elif traffic == "medium":
    ETA += 10
  elif traffic == "low":
    ETA += 5
# Weather delay
  if weather == "rainy":
    ETA += 30
  elif weather == "snowy":
    ETA += 25
  elif weather == "windy":
    ETA += 5
  print("Estimated Delivery Time:", ETA , "minutes")
#input from user
distance = int(input("Enter the distance (km): "))
weather = input("Enter the weather condition (sunny/rainy/snowy/windy):")
traffic = input("Enter the traffic level (low/medium/high): ")
delivery_time(distance, weather, traffic)
```

```python
""" Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
Build a function that analyzes seat booking data and:
Calculates occupancy percentage
Determines if show is Housefull
Suggests opening additional shows
"""
"""
Problem Statement 3: Movie Theatre Seat Occupancy Analyzer
"""
# Function to analyze seat occupancy
def seat_booking(totalseats, seats_booked):
    # Count booked seats
    countof_booked = len(seats_booked)
    # Calculate occupancy percentage
    occupied = (countof_booked / totalseats) * 100
    print("Occupancy:", int(occupied), "%")
    # Determine show status
    if occupied == 100:
        print("Show Status: Housefull")
    elif occupied >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
    # Suggest additional show
    if occupied >= 90:
        print("Suggestion: Consider opening additional show")
#input
totalseats = 200
seats_booked = [1] * 150
seat_booking(totalseats, seats_booked)
```

```python
""" Problem Statement 4:Cloud Server Load Classification System
Create a function to classify server load based on CPU usage readings.
Rules:
Average CPU < 50% → Normal
50%–80% → Warning
80% → Critical"""
#function to classify server load based on CPU usage reading
def server_load(cpu_usage):
    total = 0
    for i in cpu_usage:
        total += i
    average = int (total/len(cpu_usage))
    print("Average CPU Load:", average , "%")
    if average < 50:
        print("Server Status: Normal")
    elif average >= 50 and average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
#user input
cpu_usage = [45, 60, 70, 85, 90]
server_load(cpu_usage)
```

```python
"""Problem Statement 5:Smart Classroom Resource Usage Monitor
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns."""
"""
Problem Statement 5: Smart Classroom Resource Usage Monitor
"""
# Function to monitor resource usage
def resource_tracker(resourceUsage):
    limit = 8
    overUsed = []
    for resource, hrs in resourceUsage.items():
        if hrs > limit:
            overUsed.append(resource)
    # Output
    if overUsed:
        print("Overused Resources:", ", ".join(overUsed))
        print("Energy Alert: Yes")
    else:
        print("No overused resources")
        print("Energy Alert: No")
# Input
Usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_tracker(Usage)
```

```python
""" Problem Statement 6: Online Event Registration Capacity Controller
Create a function that manages event registrations by:
Tracking registrations
Preventing overbooking
Triggering waitlist mode"""
#function that manages event registrations
def event_register(total_capacity, registration_list):
  confirm = 0
  waiting = 0
  for x in range(registration_list):
    if x < total_capacity:
      confirm += 1
    else:
      waiting += 1
  print("Total Registrations:", confirm)
  print("Total Waiting:", waiting)
  if confirm >= total_capacity:
    print("Registration Status: Closed")
  else:
    print("Registration Status: Open")
#input
total_capacity = 100
registration_list = 105
event_register(total_capacity, registration_list)
```
